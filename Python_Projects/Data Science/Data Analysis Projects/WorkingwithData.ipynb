```python
# Data Extraction from the dataset using relational operators

maleliverdata=liver[liver['Gender']=="Male"]
print(maleliverdata.head(10))

```

```python
maleSqptlessthan50=maleabove50[maleabove50['Sqpt']<=50]
print(maleSqptlessthan50)
```

```python
# Data Extraction using Logical Operators

print(liver)
```

```python
filter1=liver[(liver['Age']>=50) & (liver['DB']<=6) & (liver['Gender']=="Female")]
print(filter1)
filter1['DB'].sum()
```

```python
filter2=liver[((liver['Gender']=="Male") & (liver['Alkphos']<=300) | (liver['Gender']=="Female") & (liver['Alkphos']<=500))]
print(filter2)
```

```python
filter3=liver[((liver['Gender']=="Male") | (liver['Alkphos']<=300) & (liver['Gender']=="Female") | (liver['Alkphos']<=500))]
print(filter3)
```
